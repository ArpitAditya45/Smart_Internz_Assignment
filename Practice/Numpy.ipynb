<h1>Data Visulization</h1>

```python
# Data Visualization
import matplotlib.pyplot as plt
```

```python
days=["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
walking_steps=[6534,7000,8900,10786,3467,11045,5095]
last_week_steps=[7890,8890,5678,9000,11000,8700,6457]
plt.plot(days,walking_steps)
plt.show()
```

```python
plt.plot(days,walking_steps,'o')
```

```python
plt.plot(days,walking_steps,linestyle='--',color='r')
```

```python
plt.plot(days,walking_steps,'o-g')
```

```python
plt.plot(days,last_week_steps,'o-m')
```

```python
plt.title("Walking steps Comparison")
plt.plot(days,walking_steps,'o-g',label='This Week')
plt.plot(days,last_week_steps,'o-m',label='Last Week')
plt.legend(loc='lower right')
plt.xlabel("Steps_walked")
plt.ylabel("days_of_the_week")
plt.show()
```

```python
plt.title("Walking steps Comparison")
plt.plot(days,walking_steps,'o-g',label='This Week')
plt.plot(days,last_week_steps,'o-m',label='Last Week')
plt.legend(loc='lower right')
plt.xlabel("Steps_walked")
plt.ylabel("days_of_the_week")
plt.grid(True)
plt.show()
```

```python
n=5
plt.figure(figsize=(n,n))#changes the size of the graph
age=[36,34,41,43,28]
name=["John","Tom","Harry","Peter","Jony"]
plt.bar(name,age)
```

```python
# Stacked Bar Chart
name=["John","Tom","Harry","Peter","Jony"]
ot=[36,34,41,43,28]
ct=[22,24,40,35,26]
plt.bar(name,ot,label="Old Timers",width=0.5,color='m')
plt.bar(name,ct,label="Current Timers",width=0.5,color='g')
plt.legend(loc='upper left')
plt.xlabel("Name")
plt.ylabel("Tickets")
plt.show()
```

```python
# pie_chart
age=[36,34,41,43,28]
name=["John","Tom","Harry","Peter","Jony"]
colors=['r','g','m','b','y']
plt.pie(age,labels=name,colors=colors,autopct='%1.1f%%',startangle=90) #autopct is used to show the percentage
plt.show()
```
