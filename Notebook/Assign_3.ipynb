## Encoding the data

<p>We perform label encoding and avoid one hot encoding as the data dimensonality will increase</p>

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
# since nan is there we need to replace it as well in sex
df['species']=le.fit_transform(df['species'])
df['island']=le.fit_transform(df['island'])
# encoding the categorical data
print(df['sex'].unique())
df['sex']=le.fit_transform(df['sex'])

```

## Correlation with target variable

```python
corr_mat=df.corr()
corr_mat['species'].sort_values(ascending=False)
```

```python
X=df.drop('species',axis=1)
y=df['species']
# After splitting 
display(X.head())
display(y.head())
```


## Scaling the data

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_Scaled=scaler.fit_transform(X)
X_Scaled=pd.DataFrame(X_Scaled,columns=X.columns)
X_Scaled.head()
```

## Splitting the data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_Scaled,y,test_size=0.2,random_state=1)
```

```python
# importing Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
pred_test=dt.predict(X_test)
pred_train=dt.predict(X_train)
```

```python
check_test=pd.DataFrame({'Actual':y_test,'Predicted':pred_test})
check_test.head()
```

```python
check_train=pd.DataFrame({'Actual':y_train,'Predicted':pred_train})
check_train.head()
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print("Accuracy score of train data is: ",accuracy_score(y_train,pred_train))
print("Accuracy score of test data is: ",accuracy_score(y_test,pred_test))
```
